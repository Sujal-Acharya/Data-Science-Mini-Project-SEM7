```python
fig1 = px.bar(
    kerala_data.groupby('crop_name')['area'].sum().reset_index().sort_values('area', ascending=False).head(10),
    x='crop_name', y='area',
    title='Top 10 Crops in Kerala by Cultivated Area',
    text_auto='.2s', color='area', color_continuous_scale='YlGn'
)
fig1.update_layout(xaxis_title='Crop', yaxis_title='Total Area (hectares)')
fig1.show()
```

```python
crop_diversity = kerala_data.groupby('district_name')['crop_name'].nunique().reset_index()
crop_diversity = crop_diversity.sort_values('crop_name', ascending=False)

fig3 = px.bar(
    crop_diversity, x='district_name', y='crop_name',
    title='Crop Diversity Across Districts in Kerala',
    color='crop_name', color_continuous_scale='Viridis', text='crop_name'
)
fig3.update_layout(xaxis_title='District', yaxis_title='Number of Unique Crops')
fig3.show()
```

```python
district_prod = kerala_data.groupby('district_name')['production'].sum().sort_values(ascending=False).reset_index()

fig = px.bar(
    district_prod, x='district_name', y='production',
    title="Total Crop Production by District in Kerala",
    color='production', color_continuous_scale='Plasma'
)
fig.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Aggregate data by district and crop
cluster_df = kerala_data.groupby(['district_name', 'crop_name'])[['area', 'production', 'yield']].mean().reset_index()

# Scale numerical columns for better clustering
scaler = StandardScaler()
scaled_features = scaler.fit_transform(cluster_df[['area', 'production', 'yield']])
```

```python
# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
cluster_df['Cluster'] = kmeans.fit_predict(scaled_features)
```

```python
from sklearn.decomposition import PCA
import plotly.express as px

# Reduce dimensions for visualization
pca = PCA(n_components=2)
pca_features = pca.fit_transform(scaled_features)
cluster_df['PCA1'] = pca_features[:,0]
cluster_df['PCA2'] = pca_features[:,1]

# Plot clusters
fig = px.scatter(
    cluster_df, x='PCA1', y='PCA2', color=cluster_df['Cluster'].astype(str),
    hover_data=['district_name', 'crop_name', 'area', 'production', 'yield'],
    title="Crop–District Clusters in Kerala Based on Area, Production & Yield",
    color_discrete_sequence=px.colors.qualitative.Bold,
    size='yield'
)
fig.show()
```

```python
cluster_summary = cluster_df.groupby('Cluster')['crop_name'].agg(lambda x: x.value_counts().index[0])
print("Dominant crop in each cluster:\n", cluster_summary)
```

```python
fig = px.bar(
    cluster_df.groupby(['Cluster', 'crop_name']).size().reset_index(name='count'),
    x='Cluster', y='count', color='crop_name',
    title="Crop Distribution Across Clusters",
    barmode='group'
)
fig.show()
```

```python
cluster_means = cluster_df.groupby('Cluster')[['area', 'production', 'yield']].mean().reset_index()
fig = px.bar(
    cluster_means.melt(id_vars='Cluster', var_name='Feature', value_name='Mean Value'),
    x='Cluster', y='Mean Value', color='Feature', barmode='group',
    title="Average Area, Production, and Yield per Cluster"
)
fig.show()
```
